## Задание 1. Колебания курса криптовалюты

Записаны изменения курса криптовалюты за 30 дней:

- Дни: [1, 2, 3, ..., 30]
- Курс ($): [50000, 50500, 52000, 51500, 53000, 52500, 54000, 54500, 55000, 55500, ..., 58000]

Используйте стиль графика с соединением точек пунктирными линиями.


```python
import matplotlib.pyplot as plt

days = list(range(0, 30))
prices = [
    50000, 50500, 52000, 51500, 53000, 52500, 54000, 54500, 55000, 55500,
    56000, 57000, 57500, 58000, 59000, 58500, 58000, 57500, 56500, 56000,
    55500, 55000, 54500, 54000, 53500, 53000, 52500, 52000, 51000, 50500
]

plt.plot(days, prices, marker='o', linestyle=':', color='b')  # Пунктир с маркерами
plt.title('Колебания курса криптовалюты за 30 дней')
plt.xlabel('Дни')
plt.ylabel('Курс ($)')
plt.grid(True)
plt.show()
```

## Задание 2. Доля пользователей разных мессенджеров

Распределение пользователей (в %):

- Мессенджеры: ['WhatsApp', 'Telegram', 'WeChat', 'Viber', 'Facebook Messenger']
- Доля (%): [40, 20, 25, 10, 5]

Постройте горизонтальную столбчатую диаграмму и выделите мессенджер с наибольшей долей.

```python
import matplotlib.pyplot as plt

messengers = ['WhatsApp', 'Telegram', 'WeChat', 'Viber', 'Facebook Messenger']
shares = [40, 20, 25, 10, 5]

colors = ['orange' if share == max(shares) else 'skyblue' for share in shares]

plt.figure(figsize=(8, 5))
bars = plt.barh(messengers, shares, color=colors, edgecolor='black')

for bar in bars:
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2, f'{width}%', va='center')

plt.title('Доля пользователей разных мессенджеров')
plt.xlabel('Доля (%)')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Задание 3 (Вариант 2, т.к. всего 20 вариантов). Площадь под синусоидой

Вычислите площадь под графиком функции:

$$ f(x) = \sin(x) $$

на интервале $[0, \pi]$.

Дополнительные требования:
- Используйте `fill_between()` для закрашивания.
- Добавьте заголовок, подписи осей.
- Отобразите оси X и Y для лучшей визуализации.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi, 500)
y = np.sin(x)

area = np.trapz(y, x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', label=r'$f(x) = \sin(x)$')

plt.fill_between(x, y, color='skyblue', alpha=0.4, label=f'Площадь ≈ {area:.4f}')

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.title(r'Площадь под графиком $f(x) = \sin(x)$ на интервале $[0, \pi]$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Задание 4. Седловая поверхность (гиперболический параболоид)

Постройте 3D-график функции:

$$ f(x, y) = x^2 - y^2 $$

на интервале $x, y \in [-5, 5]$.

Дополнительные требования:
- Используйте `plot_surface()` для отображения поверхности.
- Настройте цветовую карту через параметр `cmap`.
- Добавьте подписи осей и заголовки.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = X**2 - Y**2

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')


ax.set_title(r'Седловая поверхность $z = x^2 - y^2$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='z')

plt.tight_layout()
plt.show()
```
