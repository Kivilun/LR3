## Задание 1. Построение простого графика

Напишите код, который строит график функции $y = x^2$ на интервале \[-10, 10]\.  
Добавьте заголовок, подписи осей и сетку.

```python
import numpy as np
import matplotlib.pyplot as plt

# Задание 1. Построение простого графика
x = np.linspace(-10, 10, 400)
y = x**2

plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title('График функции $y = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## Задание 2. Настройка стилей и цветов

Постройте три линии на одном графике:

- $y = x$ (синяя, пунктирная линия),
- $y = x^2$ (зелёная, штрих-пунктирная линия),
- $y = x^3$ (красная, сплошная линия).

Добавьте легенду и сделайте оси одинакового масштаба.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)

y1 = x
y2 = x**2
y3 = x**3

plt.figure(figsize=(7, 5))

plt.plot(x, y1, linestyle='--', color='blue', label=r'$y = x$')       # синяя, пунктирная
plt.plot(x, y2, linestyle='-.', color='green', label=r'$y = x^2$')    # зелёная, штрих-пунктирная
plt.plot(x, y3, linestyle='-', color='red', label=r'$y = x^3$')       # красная, сплошная

plt.title('Графики функций $y = x$, $x^2$, $x^3$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## Задание 3. Использование различных типов графиков

Сгенерируйте 50 случайных точек и постройте диаграмму рассеяния (*scatter plot*), где цвет точек зависит от их координаты по оси $x$, а размер точек зависит от координаты по оси $y$.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Генерация 50 случайных точек
x = np.random.rand(50)
y = np.random.rand(50)

# Цвет зависит от x, размер — от y
colors = x
sizes = y * 1000

# Построение scatter plot
plt.figure(figsize=(7, 5))
scatter = plt.scatter(x, y, c=colors, s=sizes, cmap='plasma', alpha=0.7, edgecolors='k')

plt.title('Диаграмма рассеяния: цвет зависит от x, размер — от y')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(scatter, label='Значение x')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Задание 4. Гистограмма распределения

Сгенерируйте 1000 случайных чисел из нормального распределения с параметрами $\mu = 0$, $\sigma = 1$ и постройте их гистограмму с 30 бинами. Добавьте вертикальную линию в среднем значении.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

mean_val = np.mean(data)

plt.figure(figsize=(7, 5))
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Среднее = {mean_val:.2f}')

plt.title('Гистограмма нормального распределения')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Задание 5. Столбчатая диаграмма

Создайте столбчатую диаграмму, которая показывает количество студентов, получивших оценки:

- "Отлично" — 20 человек,  
- "Хорошо" — 35 человек,  
- "Удовлетворительно" — 30 человек,  
- "Неудовлетворительно" — 15 человек.

Добавьте подписи к осям и заголовок.


```python
import matplotlib.pyplot as plt

grades = ['Отлично', 'Хорошо', 'Удовлетворительно', 'Неудовлетворительно']
counts = [20, 35, 30, 15]

plt.figure(figsize=(7, 5))
bars = plt.bar(grades, counts, color='skyblue', edgecolor='black')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{height}', ha='center', va='bottom')

plt.title('Распределение студентов по оценкам')
plt.xlabel('Оценка')
plt.ylabel('Количество студентов')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Задание 7. Трёхмерный график поверхности

Используя `mpl_toolkits.mplot3d`, постройте 3D-график функции $z = \sin\left(\sqrt{x^2 + y^2}\right)$ на сетке значений $x, y$ в диапазоне $[-5, 5]$.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_title(r'$z = \sin\left(\sqrt{x^2 + y^2}\right)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.tight_layout()
plt.show()
```

## Задание 8. Множественные подграфики (subplots)

Постройте четыре графика в одной фигуре (сетка 2 × 2):

1. Линейный график $y = x$
2. Парабола $y = x^2$
3. Синус $y = \sin(x)$
4. Косинус $y = \cos(x)$

Добавьте заголовки к каждому подграфику.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, x, color='blue')
axs[0, 0].set_title(r'Линейная функция $y = x$')
axs[0, 0].grid(True)

axs[0, 1].plot(x, x**2, color='green')
axs[0, 1].set_title(r'Парабола $y = x^2$')
axs[0, 1].grid(True)

axs[1, 0].plot(x, np.sin(x), color='red')
axs[1, 0].set_title(r'Синус $y = \sin(x)$')
axs[1, 0].grid(True)

axs[1, 1].plot(x, np.cos(x), color='purple')
axs[1, 1].set_title(r'Косинус $y = \cos(x)$')
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
